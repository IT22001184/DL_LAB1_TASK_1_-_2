```python
import numpy as np
import matplotlib.pyplot as plt

#Random Distributions
exp_data = np.random.exponential(scale=1.0,size=(100000,1)) #Exponetial distribution
uniform_data = np.random.rand(100000,1) #Uniform distribution
normal_data = np.random.normal(loc=0.0,scale=1.0, size=(100000,1)) #Normal distribution

#Create the plot
plt.figure(figsize=(10,6))

#Plot histograms
plt.hist(exp_data,density=True,bins=50, histtype="step",color="orange",label="Exponential")
plt.hist(uniform_data,density=True,bins=50, histtype="step",color="blue",label="Uniform")
plt.hist(normal_data,density=True,bins=50, histtype="step",color="green",label="Normal")

plt.axis([-2.5, 8, 0, 1.5])

plt.legend(loc="upper right")
plt.title("Exponetial vs Uniform vs Normal Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)

plt.show()
```

3. Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5].

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

z = X**2 + Y**2

#3d plot creation
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, z, cmap='viridis', edgecolor='none')

ax.set_title('3D Plot of Z = X2 + Y 2')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

4. Calculate the pearson standard coefficient and spearman rank correlation for the [HP, Attack, Defense, Sp. Ark, Sp. Def, Speed] features in the seaborn tutorial.ipynb. Visualize the results using heatmaps. Make sure the values are displaced in the heatmap as well.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Pokemon dataset
!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# Relevant features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df_stats = df[features]

#Pearson and Spearman correlation matrices
pearson_corr = df_stats.corr(method='pearson')
spearman_corr = df_stats.corr(method='spearman')

# Pearson heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(pearson_corr, annot=True, cmap='Blues', fmt=".2f")
plt.title('Pearson Correlation Pokemon Stats')
plt.show()

# Spearman heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr, annot=True, cmap='Purples', fmt=".2f")
plt.title('Spearman Correlation Pokemon Stats')
plt.show()
```
