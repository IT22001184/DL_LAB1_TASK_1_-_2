4.
Plot the histogram of each of these features in a 1x4 layout.

```python
import matplotlib.pyplot as plt

selected_df = df[["N_total", "N_bulk", "N_surface", "R_avg"]]

plt.figure(figsize=(20, 4))

for i, column in enumerate(selected_df.columns):
    plt.subplot(1, len(selected_df.columns), i + 1)
    plt.hist(selected_df[column], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

5.
Visualize the scatter plots and histograms of this dataframe using the pairplot functionality of seaborn library.

```python
import seaborn as sns
import matplotlib.pyplot as plt

selected_df = df[["N_total", "N_bulk", "N_surface", "R_avg"]]

sns.pairplot(selected_df)

plt.show()
```

Add the below code and change it such that,

```python
import seaborn as sns
import matplotlib.pyplot as plt

new_df = df[["N_total", "N_bulk", "N_surface", "R_avg"]]

# PairGrid
g = sns.PairGrid(new_df)

# Upper triangle: scatter plots
g.map_upper(sns.scatterplot)

# Diagonal: histograms + KDE
g.map_diag(sns.histplot, kde=True)

# Lower triangle: bivariate KDE plots
g.map_lower(sns.kdeplot, fill= True)

plt.show()
```
